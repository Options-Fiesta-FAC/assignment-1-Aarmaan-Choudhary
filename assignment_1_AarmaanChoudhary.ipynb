```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
strike_price=float(input("Enter the strike price: "))
premium=float(input("Enter the premium: "))
spot_prices=np.linspace(strike_price-2*premium,strike_price+2*premium,100)
```

```python

fig, axes = plt.subplots(2, 2, figsize=(10, 8), dpi=100)

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Buy Call
profit_buy_call = np.maximum(0, spot_prices - strike_price) - premium
axes[0].plot(spot_prices, profit_buy_call, label="Buy Call")
axes[0].set_title("Buy Call Option")
axes[0].set_xlabel("Spot Price")
axes[0].set_ylabel("Profit")
axes[0].legend()
axes[0].grid(True)
# Sell Call
profit_sell_call = premium - np.maximum(0, spot_prices - strike_price)
axes[1].plot(spot_prices, profit_sell_call, label="Sell Call")
axes[1].set_title("Sell Call Option")
axes[1].set_xlabel("Spot Price")
axes[1].set_ylabel("Profit")
axes[1].legend()
axes[1].grid(True)

# Buy Put
profit_buy_put = np.maximum(0, strike_price - spot_prices) - premium
axes[2].plot(spot_prices, profit_buy_put, label="Buy Put")
axes[2].set_title("Buy Put Option")
axes[2].set_xlabel("Spot Price")
axes[2].set_ylabel("Profit")
axes[2].legend()
axes[2].grid(True)

# Sell Put
profit_sell_put = premium - np.maximum(0, strike_price - spot_prices)
axes[3].plot(spot_prices, profit_sell_put, label="Sell Put")
axes[3].set_title("Sell Put Option")
axes[3].set_xlabel("Spot Price")
axes[3].set_ylabel("Profit")
axes[3].legend()
axes[3].grid(True)

plt.tight_layout()
plt.show()
```

```python
#Buy call option analysis
#The buy call option remains in constant loss of 30, which is the premium and when the spot price reaches the strike price the value of the profit starts increasing and reaches 0 at the break -even point at strike price+premium
```

Buy call option analysis

The buy call option remains in constant loss of 30, which is the premium and when the spot price reaches the strike price the value of the profit starts increasing and reaches 0 at the break -even point at strike price+premium.

Sell call option analysis

The sell call option remains in the constant profit region of 30 till the spot price reaches the strike price, and then faces a decreasing trend till and breaks even at 130, finally going down to 30 loss

Buy put option analysis

The option starts with a 30 profit and breaks even at 70(strike price-premium), after which it decreases till 30 loss and then goes constant.

sell put option analysis
The options starts with a 30 loss and increases, attaining break even at 70(strike price-premium). It increases till 100(strike price) and then remains constant.
